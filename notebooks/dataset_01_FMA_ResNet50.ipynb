### Load Metadata Files
Load FMA metadata and inspect structure.

```python
import pandas as pd

tracks_csv = DATA_RAW / 'fma_metadata' / 'tracks.csv'
tracks = pd.read_csv(tracks_csv, index_col=0, header=[0, 1])

tracks.head()
```

```python
# Get subset (small/medium/large/full)
subset = tracks[('set', 'subset')]
# Get top-level genre (8 broad genres)
genre_top = tracks[('track', 'genre_top')]

meta = pd.DataFrame({
    'subset': subset,
    'genre_top': genre_top
})
meta.head()
```

```python
small = meta[meta['subset'] == 'small'].copy()
small = small[small['genre_top'].notna()]  # drop tracks with no top genre
print(len(small))
small['genre_top'].value_counts()
```

```python
small.to_csv(BASE_DIR / 'fma_small_tracks_genre_top.csv')
```

```python
from pathlib import Path

AUDIO_DIR = DATA_RAW / 'fma_small'

def track_id_to_path(track_id: int) -> Path:
    tid_str = f"{track_id:06d}"         # zero-pad to 6 digits
    folder = tid_str[:3]               # first 3 digits
    return AUDIO_DIR / folder / f"{tid_str}.mp3"

# Quick sanity check
example_id = small.index[0]
print(example_id, "->", track_id_to_path(example_id))
```

### Define Function: Audio → Mel-Spectrogram Chunks
Core preprocessing step for converting raw audio.

```python
import gc
from pathlib import Path

import librosa
import numpy as np
from PIL import Image


def audio_to_spectrogram_chunks(
    audio_path: Path,
    out_dir: Path,
    clip_duration: float = 3.0,
    sr: int = 22050,
    n_mels: int = 128,
    hop_length: int = 512,
    top_db: float = 80.0,
) -> int:
    """
    Convert an audio file into multiple Mel-spectrogram PNGs, one per fixed-length clip.

    - Uses Mel-spectrogram in dB.
    - Fixed dynamic range [-top_db, 0] dB for consistency across samples.
    - Saves as 3-channel RGB (good for ImageNet-pretrained models).
    """

    try:
        # Load mono audio at fixed sampling rate
        y, sr = librosa.load(audio_path, sr=sr, mono=True)
    except Exception as e:
        print(f"Error loading {audio_path}: {e}")
        return 0

    clip_len = int(clip_duration * sr)
    num_clips = len(y) // clip_len
    if num_clips == 0:
        return 0

    out_dir.mkdir(parents=True, exist_ok=True)
    saved = 0

    for i in range(num_clips):
        clip = y[i * clip_len : (i + 1) * clip_len]

        # Mel-spectrogram (power) -> dB
        S = librosa.feature.melspectrogram(
            y=clip,
            sr=sr,
            n_mels=n_mels,
            hop_length=hop_length,
            power=2.0,
        )
        S_db = librosa.power_to_db(S, ref=np.max, top_db=top_db)
        # S_db is now in [-top_db, 0]

        # Normalize from [-top_db, 0] -> [0, 255]
        S_norm = (S_db + top_db) / top_db  # -> [0, 1]
        S_norm = np.clip(S_norm, 0.0, 1.0)
        S_img = (S_norm * 255).astype("uint8")

        # Create image: grayscale -> RGB (3-channel for pretrained CNNs)
        img = Image.fromarray(S_img, mode="L").convert("RGB")

        fname = f"{audio_path.stem}_chunk{i:02d}.png"
        out_path = out_dir / fname
        img.save(out_path)

        saved += 1

        # Help free memory in long runs
        del clip, S, S_db, S_norm, S_img, img
        gc.collect()

    del y
    gc.collect()

    return saved
```

### Generate All Spectrograms
Convert audio clips into mel-spectrogram image chunks.

```python
from tqdm.auto import tqdm

def generate_split_spectrograms(split_name: str, df: pd.DataFrame):
    print(f"=== Generating spectrograms for {split_name} "
          f"({len(df)} tracks) ===")

    total_images = 0

    for _, row in tqdm(df.iterrows(), total=len(df)):
        track_id  = int(row['track_id'])
        genre     = row['genre_top']
        audio_p   = track_id_to_path(track_id)
        out_dir   = DATA_SPEC / split_name / genre

        if not audio_p.exists():
            # some tracks can be missing; FMA warns about this
            # just skip them
            continue

        total_images += audio_to_spectrogram_chunks(audio_p, out_dir)

    print(f"{split_name}: saved {total_images} spectrogram images.")


generate_split_spectrograms('train', train_df)
generate_split_spectrograms('valid', valid_df)
generate_split_spectrograms('test',  test_df)
```
